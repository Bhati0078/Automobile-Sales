```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('/content/drive/MyDrive/Colab Notebooks/automobile_data.csv')
df.head()
```

```python
df[df['normalized-losses']== 'NaN']
```

```python
# Null Values In DataFrame
df[df.isna()].sum()
```

```python
#Shape of DataFrame
df.shape
```

```python
# replace all ? with NAN
df.replace('?',np.nan,inplace=True)
df
```

```python
# convert the alphabet number
df['num-of-doors'].replace({'two':2,'four':4,'?':0},inplace=True)
df['num-of-doors'].unique()
```

```python
# Convert the Data Type Of Price Column object to float
df['price'] = df['price'].astype(float)
df.info()
```

```python
# Print the company name which manufactured the most expensive car and price of car.

# Find the row with the highest price
most_expensive_car = df[df['price'] == df['price'].max()]

# Extract the company name and price
company_name = most_expensive_car['make'].values[0]
price = most_expensive_car['price'].values[0]

print(f"The most expensive car is manufactured by {company_name} with a price of ${price}")

```

```python
# Print the company name which manufactured the most expensive car and price of car.
df.loc[df['price']== df['price'].max(),['make','price']]
```

```python
# Other way To get most expensive car
pd.DataFrame(df.groupby(['make'])['price'].max().sort_values(ascending=False)[0:1])
```

```python
# calculate maximum horsepowers for each company.Display group by make and max hoursepower.
pd.DataFrame(df.groupby('make')['horsepower'].max())
```

```python
# print the total no. of cars manufactured by each company.
#using pd.DataFrame to get Result in DataFrame
pd.DataFrame(df['make'].value_counts())
```

```python
# create a new columns for the updted price.this updated price will be calculate on top of the given price IF the engine in the front,price will be same else if the engine is in rear , price will be doubled.
df['Update_price']=np.where(df['engine-location']== 'front',df['price'],df['price']*2)
df
```

```python
# sort the dataframe by car and price columns.
df.sort_values(by=['make', 'price'], ascending=[True, True])
df
```

```python
# convert the alphabet number
df['num-of-doors'].unique()
df['No. of doors'] = np.where(df['num-of-doors']=='two',2,np.where(df['num-of-doors']=='four',4,np.where(df['num-of-doors']=='nan',0,df['num-of-doors'])))
df['No. of doors'].unique()
```
